```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=[1,3,6,7,9,11,15]
y=[2,3,6,8,10,13,16]
```

## Line plot

```python
plt.plot(x,y,marker="o",color="r",linestyle="-." ,label="my first line")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line plot")
plt.legend(loc="best")
plt.grid()
```

```python
plt.figure(figsize=(10,5))
plt.plot(x,y,marker="o",color="r",linestyle="-." ,label="red line")
plt.plot(y,x,marker="*",color="g",label="green line")
plt.title("line plot")
plt.legend()
```

## Scatter plot

```python
plt.scatter(x,y,s=100,color="y",label="point 1")
plt.legend()
```

```python
plt.scatter(x,y,s=200,color="y",label="point 1")
plt.scatter(y,x,s=100,label="point 2")
plt.legend()
```

## bar chart

```python
plt.bar(x,y,color=["r","g","b"])
plt.xlabel(" x axis")
```

## histogram

```python
z=[1,1,2,3,4,5,3]
plt.hist(z)
plt.xlabel('Element')
plt.ylabel("Frequency")
```

## box plot

```python
plt.boxplot(z)
```

## Pie chart

```python
plt.pie([20,70,10],labels=["a","b","c"],autopct="%1.3f%%",shadow=True);
```
